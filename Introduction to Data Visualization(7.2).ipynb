```python
# Changing the column names
```

```python
colnames =['custid', 'sex', 'employed', 'income',
           'mstat', 'hinsurance',
       'htype', 'recentm', 
           'nvehicles', 'age', 'stateofres']
```

```python
df.columns=colnames
```

```python
df.isna().sum() #Finding the total number of na's from each of the columns
```

```python
df['employed']  = df['employed'].fillna(df['employed'].mode())

df['htype']  = df['htype'].fillna(df['htype'].mode())

df['recentm']  = df['recentm'].fillna(df['recentm'].mode())
```

## Bivariate Plots

This type of plots is used when you need to find a relation between two variables and to find how the value of one variable changes the value of another variable. Different types of plots are used based on the data type of the variable.

```python
sns.relplot(x="age", y="income", data=df);
```

```python
sns.relplot(x="age", y="income",hue="sex", data=df);
```
